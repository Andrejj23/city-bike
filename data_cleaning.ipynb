```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/Trips_2018/Trips_2018.csv')
```

```python
data.head()
```

Drop Unnamed (id Collumn)

```python
data = data.drop('Unnamed: 0',axis=1)
```

Configure datetime to our date data

```python
data['start_time'] = pd.to_datetime(data['starttime'], format="%Y-%m-%d %H:%M:%S.%f")
data['stop_time'] = pd.to_datetime(data['stoptime'], format="%Y-%m-%d %H:%M:%S.%f")
```

Drop old date collumns

```python
data = data.drop('starttime',axis=1)
data = data.drop('stoptime',axis=1)
```

Ensure latitude/longitude values fall within New York City bounds.


```python
before = len(data)
data = data[(data["start_station_latitude"].between(40.45, 40.92)) &
            (data["start_station_longitude"].between(-74.30, -73.65))]
after = len(data)

print("Rows before filtering:", before)
print("Rows after filtering:", after)
print("Rows removed:", before - after)
```

Remove Tuples that have Missing Values

```python
before = len(data)

data = data.dropna()

after = len(data)

print("Rows before dropna:", before)
print("Rows after dropna:", after)
print("Rows removed by dropna:", before - after)
print(f"Removed {(before - after) / before * 100:.2f}% of total rows")
```

```python
data["usertype"].unique()
```

```python
data["usertype"] = data["usertype"].map({
    "Subscriber": 1,
    "Customer": 0
})
```
